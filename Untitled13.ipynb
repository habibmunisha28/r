```python
import pandas as pd
```

```python
df=pd.read_csv("/content/data_YesBank_StockPrices.csv")
```

```python
df
```

```python
# prompt: linear regression graph

import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

# Assuming 'Close' is the column you want to use for regression
X = np.array(df.index).reshape(-1, 1)  # Use index as the independent variable
y = df['Close']

# Create and fit the linear regression model
model = LinearRegression()
model.fit(X, y)

# Make predictions
y_pred = model.predict(X)

# Plot the data and the regression line
plt.scatter(X, y, color='blue', label='Actual Data')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Index')
plt.ylabel('Close Price')
plt.title('Linear Regression of Close Price')
plt.legend()
plt.show()
```
